```python
# Import required libraries 
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Create a helper function to display an image 
def display_image(title="", image=None, size=10):
  """
  This function takes an image as an input and returns a displayed image.
  """

  w, h = image.shape[0], image.shape[1]
  aspect_ratio = w/h 
  plt.figure(figsize=(size * aspect_ratio, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show() # display the image
```

#### **Convert a color image to grayscale**
To convert an Image to grayscale we need to use `cvtColor` function. 
It takes 2 arguments, 
the first being the input image
the second being the color space conversion code 

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
display_image("Grayscale Image", gray_image)
```

**Note**:  RGB color images have 3 dimensions, one for each primary color. 
Grayscale on the other side has only 1, which is the intensity of gray.
